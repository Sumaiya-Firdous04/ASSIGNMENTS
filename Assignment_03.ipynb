```python
#Descriptive Analytics and Data Preprocessing on Sales & Discounts Dataset
```

```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\sdsum\\OneDrive\\Desktop\\EXCELR\\ASSIGNMENTS\\CSV FILES\\3 sales_data_with_discounts.csv")
```

```python
df.select_dtypes(include=['int64', 'float64']).columns
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64'])

mean = num_cols.mean()
median = num_cols.median()
mode = num_cols.mode().iloc[0]
std = num_cols.std()
desc_stats = pd.DataFrame({
    'Mean': mean,
    'Median': median,
    'Mode': mode,
    'Std Deviation': std
})
print(desc_stats)
```

```python
#interpretation of these statistics:

# Volume: Average sales volume is 5.07, showing moderate variation across transactions.
# Avg Price: Mean price is much higher than the median, indicating a few high-priced items.
# Total Sales Value: High mean and deviation suggest large differences in total sales amounts.
# Discount Rate (%): Discounts are fairly consistent with little variation.
# Discount Amount: A few transactions received very high discounts, causing high variability.
# Net Sales Value: Net sales values vary widely due to a few high-value transactions.
```

```python
#----------------------------------------------Data Visualization--------------------------------------------------------------
```

```python
#---Histogram------
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = ['Volume', 'Avg Price', 'Total Sales Value',
            'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']

plt.figure(figsize=(12, 10))
for i, col in enumerate(num_cols):
    plt.subplot(3, 2, i + 1)
    sns.histplot(df[col], kde=True, bins=20, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Volume: Right-skewed distribution with a few extreme high values, indicating occasional bulk sales.

# Average Price: Highly right-skewed with some very high prices acting as outliers.

# Total Sales Value: Strong right skew; few high-value transactions dominate total sales.

# Discount Rate: Shows a bimodal distribution, suggesting different discount slabs.

# Discount Amount: Right-skewed with a few unusually large discounts.

# Net Sales Value: Right-skewed with extreme values representing high-revenue transactions.

# Overall Most variables are right-skewed with outliers, indicating the need for data transformation or outlier handling.
```
