```python
# Exercise 1: Prime Numbers
# Write a Python program that checks whether a given number is prime or not. 
```

```python
n = int(input("Enter a number"))

for i in range(2,n):
    if n%i == 0:
        print("It is not a prime number")
        break
else:
    print("It is a prime number")
```

```python
# Exercise 2: Product of Random Numbers
# Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. 
# The program should then check if the user's answer is correct and display an appropriate message.
```

```python
import random
```

```python
num1 = random.randint(1,100)
num2 = random.randint(1,100)

print(f"What is the product of {num1} and {num2}?")
user_ans = int(input("Enter your answer: "))

correct_ans = num1 * num2

if user_ans == correct_ans:
    print("Well done! Your answer is correct")
else:
    print(f" Wrong! The correct answer is {correct_ans}.")
```

```python
# Exercise 4:
# ✅ Problem Statement (Question): Hospital Billing System
# A hospital charges its patients based on the following:
# •	Room charges per day: ₹2,000
# •	Doctor consultation fee (flat): ₹1,500
# •	Lab test charges: Based on the number of tests taken. Each test costs ₹300
# •	Medicine charges: Total cost of medicines provided
# •	Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied
# Write a Python program to calculate the final bill for a patient who:
# •	Stayed for 4 days
# •	Had 3 lab tests
# •	Had medicine charges of ₹2,400
```

```python
room_per_day_charge = 2000
doctor_fee = 1500
lab_fee = 300
discount_threshold = 10000
discount_rate = 0.10

days_stayed = 4
num_lab_tests = 3
medicine_charges = 2400

room_charges = room_per_day_charge * days_stayed
lab_charges = lab_fee * num_lab_tests

total_before_discount = room_charges + doctor_fee + lab_fee + medicine_charges

if total_before_discount > discount_threshold:
    discount = total_before_discount * discount_rate
else:
    discount = 0

final_bill = total_before_discount - discount 

print("Hospital Bill Summary ")
print(f"Room Charges for {days_stayed} days: ₹{room_charges}")
print(f"Doctor Consultation Fees: ₹{doctor_fee}")
print(f"Lab Test Charges {num_lab_tests} tests: ₹{lab_charges}")
print(f"Medicine Charges: ₹{medicine_charges}")
print("---------------------------------")
print(f"Total Before Discount: ₹{total_before_discount}")
print(f"Discount Applied: ₹{discount}")
print(f"Final Bill Amount: ₹{final_bill}")
```
