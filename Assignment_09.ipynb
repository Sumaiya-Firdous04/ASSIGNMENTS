```python
# Step 3 — Impute Missing Values (Best Practices)
```

```python
from sklearn.impute import SimpleImputer

num_cols = df.select_dtypes(include=['int64','float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

# Numerical imputation
num_imputer = SimpleImputer(strategy='median')
df[num_cols] = num_imputer.fit_transform(df[num_cols])

# Categorical imputation
cat_imputer = SimpleImputer(strategy='most_frequent')
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])

df.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

num_cols = df.select_dtypes(include=['int64','float64']).columns

# Standard Scaling
std = StandardScaler()
df_std = df.copy()
df_std[num_cols] = std.fit_transform(df_std[num_cols])

# Min-Max Scaling
mm = MinMaxScaler()
df_mm = df.copy()
df_mm[num_cols] = mm.fit_transform(df_mm[num_cols])

df_std.head(), df_mm.head()
```

```python
# Standard Scaling (Z-score)
# Used when:
# Data is normally distributed
# Models assume Gaussian distribution
# Algorithms: SVM, Logistic Regression, Linear Regression, PCA

# Min-Max Scaling
# Used when:
# Algorithms use distance
# Inputs need 0–1 range
# Algorithms: KNN, K-Means, Neural Networks
```

```python
# 3. FEATURE ENGINEERING
# ●	Create at least 2 new features that could be beneficial for the model. Explain the rationale behind your choices
```

```python
# Feature 1 — Age Group

df["age_group"] = pd.cut(
    df["age"],
    bins=[0, 25, 45, 65, 100],
    labels=["Youth", "Adult", "Middle-Age", "Senior"]
)

df[["age", "age_group"]].head()
```

```python
# New Feature 1
# Created by binning age into categories: Youth, Adult
# Converts continuous age values into meaningful life-stage groups, helping the model capture
```

```python
# Feature 2 — Hours Category

df["hours_category"] = df["hours_per_week"].apply(
    lambda x: "Part-time" if x < 30 else ("Full-time" if x <= 50 else "Overtime")
)

df[["hours_per_week", "hours_category"]].head()
```

```python
# New Feature 2
# Created from hours_per_week
# Converts numerical hours into meaningful work categories and captures non-linear patterns.
```

```python
# ●	Apply a transformation (e.g., log transformation) to at least one skewed numerical feature and justify your choice.

import numpy as np

df["capital-gain-log"] = np.log1p(df["capital_gain"])
df["capital-loss-log"] = np.log1p(df["capital_loss"])

df[["capital_gain", "capital-gain-log"]].head()
```

```python
# capital_gain and capital_loss are highly skewed, with many zero values and a few very large values.

# The log transformation compresses large values and reduces skewness.
# log1p is used instead of log to safely handle zero values (log(0) is undefined).
# This makes the feature more suitable for machine learning models, especially linear and distance-based models.
```
