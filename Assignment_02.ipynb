```python
# 6.	Given a 'marks' variable, print the grade:
# - 90 and above: 'A'
# - 75–89: 'B'
# - 50–74: 'C'
# - Below 50: 'Fail'
```

```python
marks = int(input("Enter your marks: "))

if marks >= 90:
    grade = 'A'
elif marks >= 75 and marks <= 89:
    grade = 'B'
elif marks >= 50 and marks <= 74:
    grade = 'C'
else:
    grade = 'Fail'

print("Your Grade is:", grade)
```

```python
# 7.	Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.
```

```python
num = int(input("Enter a number: "))

if num > 0:
    if num % 2 == 0:
        print(f"{num} is positive and even.")
    else:
        print(f"{num} is positive but odd.")
elif num == 0:
    print("The number is zero.")
else:
    print(f"{num} is negative.")
```

```python
# 11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
# - Use .info(), .describe()
# - Select 'Name' and 'City' columns
# - Drop 'City' column
# - Fill any missing values in 'Salary' column with the mean
# - Remove any duplicate rows
```

```python
data = {
    'Name': ['Aisha', 'Rahul', 'Neha', 'Arjun', 'Fatima', 
             'Ravi', 'Raju', 'Karan', 'Meena', 'Ravi'],
    'Age': [25, 30, 22, 28, 35, 26, 22, 31, 29, 26],
    'City': ['Delhi', 'Mumbai', 'Chennai', 'Kolkata', 'Hyderabad',
             'Delhi', 'Chennai', 'Pune', 'Kolkata', 'Delhi'],
    'Salary': [45000, 52000, np.nan, 48000, 55000, 50000, 52000, np.nan, 47000, 50000]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df, "\n")
```

```python
# Display info and statistics
print("DataFrame Info:")
print(df.info(), "\n")
```

```python
print("Descriptive Statistics:")
print(df.describe(), "\n")
```

```python
#Select 'Name' and 'City' columns
print("Selected Columns (Name and City):")
print(df[['Name', 'City']], "\n")
```

```python
# Drop the 'City' column
df_dropped = df.drop('City', axis=1)
print("DataFrame after dropping 'City' column:\n", df_dropped, "\n")
```

```python
# Fill missing values in 'Salary' column with the mean
mean_salary = df['Salary'].mean()
df['Salary'] = df['Salary'].fillna(mean_salary)
print(f"Mean Salary = {mean_salary:.2f}")
print("DataFrame after filling missing salaries:\n", df, "\n")
```

```python
# Remove duplicate rows
df_unique = df.drop_duplicates()
print("DataFrame after removing duplicates:\n", df_unique)
```
